## Model Application
The models that are applied are:
* Linear Regression

### Metrics Function

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
def evaluation(true, prediction):
    r2 = r2_score(true, prediction)
    mae = mean_absolute_error(true, prediction)
    mse = mean_squared_error(true, prediction)
    return r2, mae, mse
```

### 1. Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

lin_reg_model = LinearRegression()
```

```python
# Training the Model
lin_reg_model.fit(x_train, y_train)
```

```python
# Making Predictions on the test Dataset 
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
lin_reg_train_pred = lin_reg_model.predict(x_train)

print('R2 score', 'MAE', 'MSE', sep='\t')
print(list(evaluation(y_train, lin_reg_train_pred)))
```

```python
# making Predictions on the test data
lin_reg_test_pred = lin_reg_model.predict(x_test)
# lin_reg_r2_score = r2_score(y_train, lin_reg_train_pred)
# lin_reg_mae = r2_score(y_train, lin_reg_train_pred)
# lin_reg_mse = r2_score(y_train, lin_reg_train_pred)

print('R2 score', 'MAE', 'MSE', sep='\t')
print(list(evaluation(y_test, lin_reg_test_pred)))
```
